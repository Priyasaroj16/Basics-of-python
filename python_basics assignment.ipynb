 **3- Compare and contrast mutable and immutable objects in Python with examples.**

  **Ans:-**  **Key Differences:**



## **Features** ------------------------**Mutable objects**-------------------------**Immutable objects**                 



---



      Modifiability            Can be changed after creation                Cannot be changed after creation

---


  
      Memory                More efficient, no new objects for changes       May use more memory, new objects for changes

---



      I/d Memory              Remains the same when modified                 Changes when a new value is assigned

---



      Examples                   Lists, dictionaries,sets                    Strings, tuples, frozensets, numbers                          


---


  
      


```python

 # Example of mutable object (list)

     my_list = [1, 2, 3]
     my_list.append(4)  # Modifies the original list
     print(my_list)  # Output: [1, 2, 3, 4]

# Example of immutable object (string)

     my_string = "hello"
     new_string = my_string.upper()  # Creates a new string
     print(my_string)  # Output: hello
     print(new_string)  # Output: HELLO
```

 **4- Discuss the different types of operators in Python and provide examples of how they are used.**
   
 **Ans:-** In Python, operators are special symbols or keywords that carry out
  operations on values and python variables.
  
 ###  **Types of Python Operators:-**
  
###    **1-Arithmetic Operators**
     
     .Mathematical operations including addition,
      subtraction, multiplication, and division are commonly carried out
      using Python arithmetic operators.

     Eg.-        Addition           +
                 Subtraction        -
                 Division           /
                 Multiplication     *
                 Modulus            %
    
                         
###     **2-Comparison (Relational) Operators**

     .To compare two values, Python comparison operators are needed.
     .Based on the comparison, they produce
      a Boolean value (True or False).

      Eg.-        Equal                    ==
                  Not equal                !=
                  Greater than             >
                  Less than                <
                  Greater than or equal    >=
                  Less than or equal       <=

     
###     **3-Assignment Operators**

     .Python assignment operators are used to assign
      values to variables in Python and The single equal symbol (=) is
      the most fundamental assignment operator.

      Eg.-       Assignment operator         =
                 Addition assignment         +=
                 Subtraction Assignment      -=
                 Multiplication Assignment   *=
                 Division Assignment         /=
                 Remainder Assignment        %=
                 Exponent Assignment         **=
                 Floor Division Assignment   //=

###    **4-Logical Operators**

      .Python logical operators are used to compose Boolean expressions and
       evaluate their truth values.
      .They are required for the creation of conditional statements as well as
       for managing the flow of execution in programs.
      .Python has three basic logical operators: AND, OR, and NOT.


###     **5-Bitwise Operators**

     .Python bitwise operators execute operations on
      indivitual bits of binary integers.
     .They work with integer binary representations, performing logical
      operations on each bit location.

###    **6-Membership Operators**
       
      .The Python membership operators test for the membership of an object
       in a sequence, such as strings, lists, or tuples. Python offers two
       membership operators to check or validate the membership of a value.

###     **7-Identity Operators**

      .The Python Identity Operators are used to compare the
       objects if both the objects are actually of the same data
       type and share the same memory location.
      

```python
#Example for arithmetic operators
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
Addition: 13
Subtraction: 7
Multiplication: 30
Division: 3.3333333333333335
Floor Division: 3
Modulus: 1
Exponentiation: 1000

#Example for assingnment operator
x = 5
x += 3
print("x after += 3:", x)
x -= 2
print("x after -= 2:", x)
x *= 4 # x = x * 4
print("x after *= 4:", x)
x /= 6 # x = x / 6
print("x after /= 6:", x)
x after += 3: 8
x after -= 2: 6
x after *= 4: 24
x after /= 6: 4.0

#Example for Comparison (Relational) Operators
a = 5
b = 10
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= 5:", a >= 5)
print("b <= 9:", b <= 9)
a == b: False
a != b: True
a > b: False
a < b: True
a >= 5: True
b <= 9: False

#Example for Logical Operators
x = 7
print("x > 5 and x < 10:", x > 5 and x < 10)
print("x < 5 or x > 10:", x < 5 or x > 10)
print("not(x == 7):", not(x == 7))
x > 5 and x < 10: True
x < 5 or x > 10: False
not(x == 7): False

#Example for Bitwise Operators
a = 5
b = 3
print("a & b:", a & b)
print("a | b:", a | b)
print("a << 1:", a << 1)
print("a >> 1:", a >> 1)
a & b: 1
a | b: 7
a << 1: 10
a >> 1: 2

#Example of membership operators
list1 = [1, 2, 3, 4, 5]
string1 = "Hello"
tuple1 = (6, 7, 8)
set1 = {9, 10, 11}
dict1 = {"a": 1, "b": 2}

print(3 in list1)  # Output: True
print("e" in string1)  # Output: True
print(9 not in tuple1)  # Output: True
print(10 in set1)  # Output: True
print("c" not in dict1) # Output: True

#Example of identity operators
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)        # Output: True (x and y refer to the same object)
print(x is z)        # Output: False (x and z are different objects)
print(x is not z)    # Output: True (x and z are not the same object)
```

**5- Explain the concept of type casting in Python with examples.**

  **Ans:-**
      Type Casting in Python (Implicit and Explicit) with Examples:-
      
     .Type Casting is the method to convert the Python variable datatype into  
      a certain data type in order to perform the required operation by users.
      In this article, we will see the various techniques for typecasting.
      There can be two types of Type Casting in Python:

      (a)  Implicit type casting
      (b)  Explicit type casting

```python
# implicit type casting,
 x = 10  # integer
 y = 5.5 # float

result = x + y
print(result)
print(type(result))
# Expected Output:
# 15.5
# <class 'float'>

# explicit type casting,

# Example 1: Converting string to integer
string_num = "123"
int_num = int(string_num)
print(f"Type of string_num: {type(string_num)}, value: {string_num}")
print(f"Type of int_num: {type(int_num)}, value: {int_num}")

# Example 2: Converting float to integer
float_num = 3.14
int_num = int(float_num)
print(f"Type of float_num: {type(float_num)}, value: {float_num}")
print(f"Type of int_num: {type(int_num)}, value: {int_num}")

# Example 3: Converting integer to float
int_num = 10
float_num = float(int_num)
print(f"Type of int_num: {type(int_num)}, value: {int_num}")
print(f"Type of float_num: {type(float_num)}, value: {float_num}")

# Example 4: Converting to string
num = 100
string_num = str(num)
print(f"Type of num: {type(num)}, value: {num}")
print(f"Type of string_num: {type(string_num)}, value: {string_num}")

# Example 5: Converting to boolean
num = 0
bool_value = bool(num)
print(f"Type of num: {type(num)}, value: {num}")
print(f"Type of bool_value: {type(bool_value)}, value: {bool_value}")

num = 42
bool_value = bool(num)
print(f"Type of num: {type(num)}, value: {num}")
print(f"Type of bool_value: {type(bool_value)}, value: {bool_value}")

```
